```python
#merging and joins
dataf=pd.DataFrame({
    "employee":["a","b","c","a"],
    "salary":[1000,2000,3000,1000]
})
dataf
```

```python
data2=pd.DataFrame({
    "employee":["a","d","c","a"],
    "age":["10","6","3","5"]
})
data2
```

```python
ab=pd.merge(dataf,data2, on="employee")
ab
```

```python
#fitering for duplicate values
#by default subset value is none

ab.drop_duplicates(subset=None, keep="first", inplace=True)
ab
```

```python
pd.merge(dataf,data2, on="employee", how="outer")
```

```python
pd.merge(dataf,data2, on="employee",how="left")
```
